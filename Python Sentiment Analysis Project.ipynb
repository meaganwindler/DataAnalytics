```python
# Task 1: Read the CSV File, remove empty rows, and limit to 5,000 records
file_path = r"C:\Users\Mwindler\Downloads\productreviews.csv"
df = pd.read_csv(file_path)
df = df.dropna().reset_index(drop=True)
df = df.head(5000)

# Print column names to verify
print("Column names in DataFrame:", df.columns)
```

```python
# Task 2: Function to Analyze Sentiment
def analyze_text_sentiment(text):
    creds = service_account.Credentials.from_service_account_file(r"C:\Users\Mwindler\Downloads\osu-programming-assignment-8-dad3bf9f7f96.json")
    client = language.LanguageServiceClient(credentials=creds)
    document = language.Document(content=text, type_=language.Document.Type.PLAIN_TEXT)
    response = client.analyze_sentiment(document=document)
    sentiment = response.document_sentiment

    # Print statements
    print(text)
    print("score: " + str(sentiment.score))
    print("magnitude: " + str(sentiment.magnitude))

    return sentiment.score, sentiment.magnitude
```

```python
# Analyzing Sentiment for Each Review with Sleep After Every 500 Rows
scores = []
magnitudes = []
for index, row in df.iterrows():
    score, magnitude = analyze_text_sentiment(row['Review'])
    scores.append(score)
    magnitudes.append(magnitude)
    df.at[index, 'sentiment_score'] = score
    df.at[index, 'magnitude'] = magnitude

    if (index + 1) % 500 == 0:
        time.sleep(60)  # Sleep for 60 seconds
```

```python
# Task 3: Save the Updated DataFrame
df.to_csv(r"C:\Users\Mwindler\Downloads\productreviewssentiment.csv", index=False)
```
